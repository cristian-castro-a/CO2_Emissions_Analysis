**Goal**:
- Forecast the amount of World CO2 emissions (in tonnes)

**Comments:**
- Column 'CO2' of the dataframe indicates the Tonnes of CO2 emmitted into the atmosphere.

```python
# Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from prophet import Prophet
```

## Data

```python
df = pd.read_csv('owid-co2-data.csv')
df.head()
```

```python
df.keys()
```

```python
df = df.groupby(by=['year']).sum().round(1).reset_index()[['year','co2']]
df.insert(loc = 1, column = 'month', value = 12)
df.insert(loc = 2, column = 'day', value = 31)
values = pd.to_datetime(df[['year','month','day']])
df.insert(loc = 0, column = 'DS', value = values)
df.drop(['year','month','day'], axis = 1, inplace = True)
df.head()
```

```python
df.isna().sum()
```

```python
# Plotting trend
plt.figure(figsize = (15,10))
plt.plot(df['DS'],
         df['co2'],
         linewidth = 3)
plt.title('Total World CO2 Emissions', fontsize = 22)
plt.xlabel('Year', fontsize = 18)
plt.ylabel('Tonnes of CO2 Emmitted into the Atmosphere', fontsize = 18)
plt.xticks(fontsize = 13)
plt.yticks(fontsize = 13)
plt.show()
```

## Prophet Model (Facebook)

```python
# Prepare the dataset
df_prophet = df.copy(deep = True)
df_prophet.rename(columns={"DS": "ds", "co2": "y"}, inplace = True)
df_prophet.head()
```

```python
# Train the model
m_prophet = Prophet(interval_width=0.95)
m_prophet.fit(df_prophet)
future = m_prophet.make_future_dataframe(periods=10, freq = 'Y')
future.tail()
```

```python
# Forecasting
forecast_prophet = m_prophet.predict(future)
forecast_prophet[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].tail()
```

```python
# Plot the forecast
fig1 = m_prophet.plot(forecast_prophet)
```

```python
# Components
fig2 = m_prophet.plot_components(forecast_prophet)
```
